```python
import math
import sys

##################################################################################################################
# Question 1

BLACK = 'B'
RED = 'R'
nil = None

def getColor(n):
    return BLACK if n is nil else n.color

def getSize(n):
    return 0 if n is nil else n.size

def is_nil(node):
    return node == nil

class RB_Node:
    def __init__(self, key):
        self.left = nil  # left child of node
        self.right = nil  # right child of node
        self.p = nil  # parent of node
        self.color = RED  # color of node
        self.key = key  # key of node
        self.size = 1  # size of the subtree of the node

    # Complete this method to calculate the size of the subtree for the node
    def sizeOf(self):
        if is_nil(self):
            return 0
        if is_nil(self.left):
            size_left = 0
        else:
            size_left = self.left.sizeOf()
        if is_nil(self.right):
            size_right = 0
        else:
            size_right = self.right.sizeOf()
        
        return size_left + 1 + size_right
    
        pass

##################################################################################################################
# Question 2

class RB_Tree:
    def __init__(self):
        self.root = nil

    def RB_INSERT(self, z):
        self.TREE_INSERT(z)
        self.RB_INSERT_FIX(z)

    # Compete the method to insert RB_node z into the tree.
    # the process is the same as BST INSERT one.
    # you can choose to use iteration or recursion.
    def TREE_INSERT(self, z):
        
        self.TREE_INSERT_node(self.root, z)
        
        
    def TREE_INSERT_node(self, node, z):
        if is_nil(self.root):
            self.root = z
            
        elif z.key <= node.key:
            if is_nil(node.left):
                node.left = z
                z.p = node
            else:
                self.TREE_INSERT_node(node.left, z)
        else:
            if is_nil(node.right):
                node.right = z
                z.p = node
            else:
                self.TREE_INSERT_node(node.right, z)
                
##################################################################################################################
# Question 3

    def LEFT_ROTATE(self, x):
        # x is the right child of its parent.
        # now you need to do the ‘rotate’ operation

        parent = x.p
        right = x.right

        x.right = right.left
        if not is_nil(x.right):
            x.right.p = x

        right.left = x
        x.p = right

        right.p = parent
        if is_nil(parent):
            self.root = right
        elif x == parent.left:
            parent.left = right
        else:
            parent.right = right



    def RIGHT_ROTATE(self, x):
        # x is the left child of its parent.
        # now you need to do the ‘rotate’ operation

        parent = x.p
        left = x.left

        x.left = left.right
        if not is_nil(x.left):
            x.left.p = x

        left.right = x
        x.p = left

        left.p = parent
        if is_nil(parent):
            self.root = left
        elif x == parent.left:
            parent.left = left
        else:
            parent.right = left

##################################################################################################################
# Question 4

    def RB_INSERT_FIX(self, z):
        # z is the current node we are dealing with.
        # from the leaves to the root of the Tree.
        while getColor(z) == RED and getColor(z.p) == RED:
            # Case 1: z's parent is a left child of z's grandparent
            if z.p == z.p.p.left:
                y = z.p.p.right 
                # Case 1a: y is red
                if getColor(y) == RED:
                    z.p.color = BLACK  
                    y.color = BLACK  
                    z.p.p.color = RED 
                    z = z.p.p
                else:
                    # Case 1b: y is black
                    if z == z.p.right:
                        z = z.p 
                        self.LEFT_ROTATE(z) 
                    # Case 1c: y is black (z is left child of its parent)
                    z.p.color = BLACK 
                    z.p.p.color = RED
                    self.RIGHT_ROTATE(z.p.p)  
            else:
                # Case 2: z's parent is a right child of z's grandparent (symmetric to case 1)
                y = z.p.p.left  
                # Case 2a: y is red
                if getColor(y) == RED:
                    z.p.color = BLACK  
                    y.color = BLACK 
                    z.p.p.color = RED  
                    z = z.p.p  
                else:
                    # Case 2b: y is black
                    if z == z.p.left:
                        z = z.p  
                        self.RIGHT_ROTATE(z) 
                    # Case 2c: y is black (z is right child of its parent)
                    z.p.color = BLACK  
                    z.p.p.color = RED  
                    self.LEFT_ROTATE(z.p.p)  
        self.root.color = BLACK  

##################################################################################################################
# Question 5

    def GET_RANK(self, val):
        # get the rank of value ‘val’
        # e.g. 5 ranks 3 in array [1,2,5,6,7], because 5 is the 3rd
        # largest number in the array.
        # using recursion, x is the current RB_node.
        # or you can delete the ‘x’ and use iteration.
        return self.GET_RANK_node(self.root, val)

    def GET_RANK_node(self, node, val):
        if is_nil(node):
            return 0
        if node.key == val:
            size_left = 0 if is_nil(node.left) else node.left.sizeOf()
            return size_left + 1
        if node.key > val:
            return self.GET_RANK_node(node.left, val)
        
        size_left = 0 if is_nil(node.left) else node.left.sizeOf()
        return size_left + 1 + self.GET_RANK_node(node.right, val)

    def GET_KEY(self, val):
        # get the value who ranks ‘val’
        # using recursion, x is the current RB_node
        # or you can delete the ‘x’ and use iteration
        return self.GET_KEY_node(self.root, val)

    def GET_KEY_node(self, node, val):
        size_left = 0 if is_nil(node.left) else node.left.sizeOf()

        if size_left == val - 1:
            return node.key
        elif size_left > val - 1:
            return self.GET_KEY_node(node.left, val)
        else:
            return self.GET_KEY_node(node.right, val - size_left - 1)


    def Successor(self, val):
        # find the minimal value larger than ‘val’ in the RB-Tree.
        # or you can delete the ‘x’ and use iteration to

        return self.Successor_node(self.root, val)


    def Successor_node(self, node, val):
        if is_nil(node):
            return math.inf
        if node.key <= val:
            return self.Successor_node(node.right, val)
        else:
            return min(node.key, self.Successor_node(node.left, val))


    def Predecessor(self, val):
        # find the maximal value less than ‘val’ in the RB-Tree
        # or you can delete the ‘x’ and use iteration to solve this

        return self.Predecessor_node(self.root, val)


    def Predecessor_node(self, node, val):
        if is_nil(node):
            return -math.inf
        if node.key >= val:
            return self.Predecessor_node(node.left, val)
        else:
            return max(node.key, self.Predecessor_node(node.right, val))

        pass
```
